```python
from google_play_scraper import app, reviews_all, Sort
import pandas as pd
import numpy as np
import csv
```

#Scraping Dataset

```python
scrap_review = reviews_all(
    'com.roblox.client',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=100000
)
```

```python
with open('reviews.csv', mode='w', encoding='utf-8') as file:
  writer = csv.writer(file)
  writer.writerow(['Review'])
  for review in scrap_review:
    writer.writerow([review['content']])
```

```python
df = pd.DataFrame(scrap_review)
df.head()
```

```python
df.to_csv('ulasan.csv', index=False)
```
